```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
def euclidian_distance(point_0, point_1):
    distance = (point_0[0] - point_1[0]) ** 2 + (point_0[1] - point_1[1]) ** 2
    return np.sqrt(distance)
```

```python
data_points = {
    "x":[],
    "y":[]
}
data_points_count = 100

for i in range(data_points_count):
    point_x = np.random.randint(1, 99)
    point_y = np.random.randint(1, 99)
    data_points["x"].append(point_x )
    data_points["y"].append(point_y )

print(data_points)
assert len(data_points["x"]) == len(data_points["y"])

centroids_count = 3
centroids = {
    "x":[],
    "y":[]
}

colors = ["y", "g", "r"]

assert len(colors) == centroids_count
for i in range(centroids_count):
    centroids["x"].append(point_x)
    centroids["y"].append(point_y)
```

```python
for i in range(centroids_count):
    plt.scatter(centroids["x"][i], centroids["y"][i], color=colors[i])
i = 0
plt.scatter(centroids["x"][i], centroids["y"][i], color=colors[i])
plt.scatter(data_points['x'], data_points['y'], color='k')
plt.xlim(0, 100)
plt.ylim(0, 100)
```

```python
centroids_position = []
for i in range(centroids_count):
    centroids_position.append([centroids["x"][i], centroids["y"][i]])
print(centroids_position)
```

```python
%matplotlib notebook
for i in range(1000):
    centroids_best_fit_point = [

    ]
    for i in range(centroids_count):
        centroids_best_fit_point.append([])
    # closest centroid from data point
    for i in range(data_points_count):
        best_fit_index = 0
        best_fit_distance = 0
        for centroid_index in range(len(centroids_position)):
            centroid = centroids_position[centroid_index]
            distance = euclidian_distance([data_points["x"][i], data_points["y"][i]], centroid)
            if(centroid_index == 0 or distance <= best_fit_distance):
                best_fit_index = centroid_index
                best_fit_distance = distance
        centroids_best_fit_point[best_fit_index].append([data_points["x"][i], data_points["y"][i]])

    for i in range(centroids_count):
        mean_x = 0
        mean_y = 0
        test = []
        for j in range(len(centroids_best_fit_point[i])):
            mean_x += centroids_best_fit_point[i][j][0]
            mean_y += centroids_best_fit_point[i][j][1]
            test.append(centroids_best_fit_point[i])

        mean_x /= len(centroids_best_fit_point[i]) + 1
        mean_y /= len(centroids_best_fit_point[i]) + 1
        
        centroids["x"][i] = mean_x
        centroids["y"][i] = mean_y
        
        centroids_position[i][0] = mean_x
        centroids_position[i][1] = mean_y 
        
#print(centroids)
for i in range(centroids_count):
    plt.scatter(centroids["x"][i], centroids["y"][i], color=colors[i])

plt.scatter(data_points['x'], data_points['y'], color='k')
plt.xlim(0, 100)
plt.ylim(0, 100)
```
