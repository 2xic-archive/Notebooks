```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

function = lambda x : (x - 2) ** 2
derivation_function = lambda x : 2*x-4

x = np.linspace(-15,15, num = 100)
y = function(x)

plt.plot(x,y)
plt.show()
```

```python
learning_rate = 0.04
iterations_rounds = 1000
preferred_precision = 0.00001
current_x  = 15
```

```python
learning_points = {
    "x":[],
    "y":[]
}
for i in range(iterations_rounds):
    old_x = current_x
    current_x -= learning_rate * derivation_function(current_x)
    if(i % 5 == 0):
        learning_points["x"].append(current_x)
        learning_points["y"].append(function(current_x))
    if(abs(current_x-old_x) < preferred_precision):
        break
    
plt.scatter(learning_points["x"], learning_points["y"])
plt.plot(x,y)
plt.show()
```
